## 3. Görüntüyü Açma
OpenCV ile bir görüntüyü açmak için cv2.imread() fonksiyonu kullanılır. Görüntü, piksel değerleri içeren bir matris olarak yüklenir.

````python
import cv2

# Görüntüyü açma
image = cv2.imread('c:\goruntu\indir.jpg')

# Görüntüyü ekranda gösterme
cv2.imshow('Görüntü', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

Açıklama: Bu blok, bir görüntüyü açma ve ekranda gösterme işlemlerini içerir.



## 4. Görüntüyü Matris Olarak Yükleme

## 5. Görüntü İşleme Temel Operasyonlar
### 5.1 Gri Tonlama

````python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gri Tonlama', gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

Açıklama: Bu blok, bir görüntüyü gri tonlama yaparak tek renkli bir görüntü elde etme işlemini gösterir.

### 5.2 Renk Uzayı Değiştirme

````python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV Renk Uzayı', hsv_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

Açıklama: Bu blok, görüntü renk uzayını değiştirerek HSV (ton, doygunluk, parlaklık) uzayına dönüştürme işlemini gösterir.

### 5.3 Eşikleme

````python
# Görüntüyü gri tonlama yapma
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Eşikleme işlemi
_, thresholded_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('Eşikleme', thresholded_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

Açıklama: Bu blok, bir görüntüyü eşikleme (thresholding) yaparak binary bir görüntü elde etme işlemini gösterir.

## 5.4 Bulanıklık (Blur)

````python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
cv2.imshow('Bulanıklık', blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

Açıklama: Bu blok, bir görüntüye Gauss filtresi uygulayarak bulanıklık efekti elde etme işlemini gösterir.
