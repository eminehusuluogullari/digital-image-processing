```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# üìå G√∂r√ºnt√ºy√º y√ºkle ve gri tonlamaya √ßevir
image = cv2.imread("C:\\Goruntu\\rices.jpg", cv2.IMREAD_GRAYSCALE)

# 1Ô∏è‚É£ **Klasik Histogram E≈üitleme**
equalized = cv2.equalizeHist(image)

# 2Ô∏è‚É£ **CLAHE (Contrast Limited Adaptive Histogram Equalization)**
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_equalized = clahe.apply(image)

# 3Ô∏è‚É£ **Histogramlarƒ± Hesapla**
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equalized], [0], None, [256], [0, 256])
hist_clahe = cv2.calcHist([clahe_equalized], [0], None, [256], [0, 256])

# üîπ **Sonu√ßlarƒ± G√∂rselle≈ütirme**
fig, axes = plt.subplots(3, 2, figsize=(15, 10))

# Orijinal G√∂r√ºnt√º
axes[0, 0].imshow(image, cmap="gray")
axes[0, 0].set_title("Orijinal G√∂r√ºnt√º")
axes[0, 0].axis("off")

# Orijinal Histogram
axes[0, 1].plot(hist_original, color='black')
axes[0, 1].set_title("Orijinal Histogram")

# Klasik Histogram E≈üitleme Sonucu
axes[1, 0].imshow(equalized, cmap="gray")
axes[1, 0].set_title("Klasik Histogram E≈üitleme")
axes[1, 0].axis("off")

# Klasik Histogram E≈üitleme Histogramƒ±
axes[1, 1].plot(hist_equalized, color='black')
axes[1, 1].set_title("Klasik Histogram E≈üitleme Histogramƒ±")

# CLAHE Histogram E≈üitleme Sonucu
axes[2, 0].imshow(clahe_equalized, cmap="gray")
axes[2, 0].set_title("CLAHE Histogram E≈üitleme")
axes[2, 0].axis("off")

# CLAHE Histogramƒ±
axes[2, 1].plot(hist_clahe, color='black')
axes[2, 1].set_title("CLAHE Histogramƒ±")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# G√∂r√ºnt√ºy√º y√ºkle
image = cv2.imread("C:\\Goruntu\\bus.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # OpenCV BGR formatƒ±nda a√ßar, RGB'ye √ßeviriyoruz

# 1Ô∏è‚É£ **Gri Tonlama (Grayscale)**
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2Ô∏è‚É£ **Yeniden Boyutlandƒ±rma (Resize)**
new_width, new_height = 300, 300
image_resized = cv2.resize(image_rgb, (new_width, new_height))

# 3Ô∏è‚É£ **D√∂nd√ºrme (Rotate)**
(h, w) = image.shape[:2]  # Y√ºkseklik ve geni≈ülik
center = (w // 2, h // 2)  # D√∂nd√ºrme merkezi
matrix = cv2.getRotationMatrix2D(center, 45, 1.0)  # 45 derece d√∂nd√ºrme
image_rotated = cv2.warpAffine(image_rgb, matrix, (w, h))

# 4Ô∏è‚É£ **Aynalama (Flip)**
image_flipped = cv2.flip(image_rgb, 1)  # 1: Yatay √ßevirme, 0: Dikey √ßevirme, -1: Her ikisi

# 5Ô∏è‚É£ **Bulanƒ±kla≈ütƒ±rma (Blur)**
image_blur = cv2.GaussianBlur(image_rgb, (15, 15), 0)

# 6Ô∏è‚É£ **Kenarlƒ±k Algƒ±lama (Canny)**
image_edges = cv2.Canny(image_gray, 100, 200)

# 7Ô∏è‚É£ **Histogram E≈üitleme (Contrast Enhancement)**
image_equalized = cv2.equalizeHist(image_gray)

# üîπ **T√ºm ƒ∞≈ülemleri G√∂r√ºnt√ºleme**
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

# Orijinal G√∂r√ºnt√º
axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title("Orijinal G√∂r√ºnt√º")
axes[0, 0].axis("off")

# Grayscale
axes[0, 1].imshow(image_gray, cmap="gray")
axes[0, 1].set_title("Gri Tonlama (Grayscale)")
axes[0, 1].axis("off")

# Yeniden Boyutlandƒ±rma
axes[0, 2].imshow(image_resized)
axes[0, 2].set_title("Yeniden Boyutlandƒ±rma (Resize)")
axes[0, 2].axis("off")

# D√∂nd√ºrme
axes[0, 3].imshow(image_rotated)
axes[0, 3].set_title("D√∂nd√ºrme (Rotate 45¬∞)")
axes[0, 3].axis("off")

# Aynalama
axes[1, 0].imshow(image_flipped)
axes[1, 0].set_title("Aynalama (Flip)")
axes[1, 0].axis("off")

# Bulanƒ±kla≈ütƒ±rma
axes[1, 1].imshow(image_blur)
axes[1, 1].set_title("Bulanƒ±kla≈ütƒ±rma (Blur)")
axes[1, 1].axis("off")

# Kenarlƒ±k Algƒ±lama
axes[1, 2].imshow(image_edges, cmap="gray")
axes[1, 2].set_title("Kenarlƒ±k Algƒ±lama (Canny)")
axes[1, 2].axis("off")

# Histogram E≈üitleme
axes[1, 3].imshow(image_equalized, cmap="gray")
axes[1, 3].set_title("Histogram E≈üitleme")
axes[1, 3].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# G√∂r√ºnt√ºy√º y√ºkle ve BGR'den RGB'ye √ßevir
image = cv2.imread("C:\\Goruntu\\bus.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 1Ô∏è‚É£ **Gaussian Blur (Gauss Bulanƒ±kla≈ütƒ±rma)**
image_gaussian = cv2.GaussianBlur(image_rgb, (15, 15), 0)

# 2Ô∏è‚É£ **Median Blur (Tuz ve Biber G√ºr√ºlt√ºs√ºne Kar≈üƒ±)**
image_median = cv2.medianBlur(image_rgb, 5)

# 3Ô∏è‚É£ **Bilateral Filter (Kenarlardan √ñd√ºn Vermeden G√ºr√ºlt√º Giderme)**
image_bilateral = cv2.bilateralFilter(image_rgb, 9, 75, 75)

# 4Ô∏è‚É£ **Sobel Kenar Algƒ±lama (X ve Y Y√∂nl√º)**
sobel_x = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=5)
sobel_combined = cv2.bitwise_or(sobel_x, sobel_y)

# 5Ô∏è‚É£ **Laplacian Kenar Algƒ±lama (ƒ∞kinci Derece T√ºrev)**
laplacian = cv2.Laplacian(image_gray, cv2.CV_64F)

# 6Ô∏è‚É£ **Canny Kenar Algƒ±lama (En Pop√ºler)**
canny_edges = cv2.Canny(image_gray, 100, 200)

# 7Ô∏è‚É£ **Prewitt Kenar Algƒ±lama (Sobel Alternatifi)**
prewitt_x = cv2.filter2D(image_gray, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(image_gray, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
prewitt_combined = cv2.bitwise_or(prewitt_x, prewitt_y)

# üîπ **T√ºm Filtreleme ve Kenar Algƒ±lama Tekniklerini G√∂rselle≈ütirme**
fig, axes = plt.subplots(3, 3, figsize=(20, 15))

# Orijinal G√∂r√ºnt√º
axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title("Orijinal G√∂r√ºnt√º")
axes[0, 0].axis("off")

# Gaussian Blur
axes[0, 1].imshow(image_gaussian)
axes[0, 1].set_title("Gaussian Blur (Bulanƒ±kla≈ütƒ±rma)")
axes[0, 1].axis("off")

# Median Blur
axes[0, 2].imshow(image_median)
axes[0, 2].set_title("Median Blur (G√ºr√ºlt√º Azaltma)")
axes[0, 2].axis("off")

# Bilateral Filter
axes[1, 0].imshow(image_bilateral)
axes[1, 0].set_title("Bilateral Filter (Kenarlarƒ± Korumalƒ± Blur)")
axes[1, 0].axis("off")

# Sobel Kenar Algƒ±lama
axes[1, 1].imshow(sobel_combined, cmap="gray")
axes[1, 1].set_title("Sobel Kenar Algƒ±lama")
axes[1, 1].axis("off")

# Laplacian Kenar Algƒ±lama
axes[1, 2].imshow(laplacian, cmap="gray")
axes[1, 2].set_title("Laplacian Kenar Algƒ±lama")
axes[1, 2].axis("off")

# Canny Kenar Algƒ±lama
axes[2, 0].imshow(canny_edges, cmap="gray")
axes[2, 0].set_title("Canny Kenar Algƒ±lama")
axes[2, 0].axis("off")

# Prewitt Kenar Algƒ±lama
axes[2, 1].imshow(prewitt_combined, cmap="gray")
axes[2, 1].set_title("Prewitt Kenar Algƒ±lama")
axes[2, 1].axis("off")

# Bo≈ü kutu (G√∂rsel uyum i√ßin)
axes[2, 2].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# üìå G√∂r√ºnt√ºy√º y√ºkle ve gri tonlamaya √ßevir
image = cv2.imread("C:\\Goruntu\\bus.jpg")
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 1Ô∏è‚É£ **Histogram Analizi (Grayscale & RGB Histogramƒ±)**
# Histogram hesaplama
hist_gray = cv2.calcHist([image_gray], [0], None, [256], [0, 256])

# RGB Histogramƒ± i√ßin kanallarƒ± ayƒ±rma
colors = ('b', 'g', 'r')  # OpenCV BGR formatƒ±nda
hist_channels = [cv2.calcHist([image], [i], None, [256], [0, 256]) for i in range(3)]

# 2Ô∏è‚É£ **Global Thresholding (Sabit E≈üikleme)**
_, thresh_binary = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

# 3Ô∏è‚É£ **Otsu Thresholding (Otomatik E≈üikleme)**
_, thresh_otsu = cv2.threshold(image_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 4Ô∏è‚É£ **Adaptive Thresholding (Uyarlanabilir E≈üikleme)**
adaptive_thresh = cv2.adaptiveThreshold(image_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# 5Ô∏è‚É£ **Binary Thresholding (ƒ∞kili E≈üikleme)**
_, thresh_binary_inv = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY_INV)

# 6Ô∏è‚É£ **Truncation (Kesme) Thresholding**
_, thresh_trunc = cv2.threshold(image_gray, 127, 255, cv2.THRESH_TRUNC)

# 7Ô∏è‚É£ **Gaussian Adaptive Thresholding (√ñnerilen Ek Teknik)**
gaussian_adaptive = cv2.adaptiveThreshold(image_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 4)

# üîπ **T√ºm ƒ∞≈ülemleri G√∂r√ºnt√ºleme**
fig, axes = plt.subplots(3, 3, figsize=(20, 15))

# Orijinal G√∂r√ºnt√º
axes[0, 0].imshow(image_gray, cmap="gray")
axes[0, 0].set_title("Orijinal Gri Tonlama G√∂r√ºnt√º")
axes[0, 0].axis("off")

# Histogram (Grayscale)
axes[0, 1].plot(hist_gray, color='black')
axes[0, 1].set_title("Grayscale Histogram")

# RGB Histogram
axes[0, 2].set_title("RGB Histogram")
for i, color in enumerate(colors):
    axes[0, 2].plot(hist_channels[i], color=color)
axes[0, 2].set_xlim([0, 256])

# Global Sabit E≈üikleme
axes[1, 0].imshow(thresh_binary, cmap="gray")
axes[1, 0].set_title("Global Binary Thresholding")
axes[1, 0].axis("off")

# Otsu Thresholding
axes[1, 1].imshow(thresh_otsu, cmap="gray")
axes[1, 1].set_title("Otsu Thresholding")
axes[1, 1].axis("off")

# Adaptive Thresholding
axes[1, 2].imshow(adaptive_thresh, cmap="gray")
axes[1, 2].set_title("Adaptive Thresholding")
axes[1, 2].axis("off")

# Binary Inverse Thresholding
axes[2, 0].imshow(thresh_binary_inv, cmap="gray")
axes[2, 0].set_title("Binary Inverse Thresholding")
axes[2, 0].axis("off")

# Truncation Thresholding
axes[2, 1].imshow(thresh_trunc, cmap="gray")
axes[2, 1].set_title("Truncation Thresholding")
axes[2, 1].axis("off")

# Gaussian Adaptive Thresholding
axes[2, 2].imshow(gaussian_adaptive, cmap="gray")
axes[2, 2].set_title("Gaussian Adaptive Thresholding")
axes[2, 2].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# üìå G√∂r√ºnt√ºy√º y√ºkleme
image = cv2.imread("C:\\Goruntu\\bus.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # OpenCV BGR formatƒ±nda a√ßar, RGB'ye √ßeviriyoruz

# 1Ô∏è‚É£ **Boyutlandƒ±rma (Resize) - B√ºy√ºtme ve K√º√ß√ºltme**
image_resized_up = cv2.resize(image_rgb, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)  # %150 b√ºy√ºtme
image_resized_down = cv2.resize(image_rgb, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)  # %50 k√º√ß√ºltme

# 2Ô∏è‚É£ **D√∂nd√ºrme (Rotate)**
(h, w) = image.shape[:2]  # Y√ºkseklik ve geni≈ülik
center = (w // 2, h // 2)  # D√∂nd√ºrme merkezi
matrix = cv2.getRotationMatrix2D(center, 45, 1.0)  # 45 derece d√∂nd√ºrme
image_rotated = cv2.warpAffine(image_rgb, matrix, (w, h))

# 3Ô∏è‚É£ **√ñl√ßekleme (Scaling)**
image_scaled = cv2.resize(image_rgb, (w // 2, h // 2), interpolation=cv2.INTER_LINEAR)

# 4Ô∏è‚É£ **Affine Transform (√áarpƒ±tma)**
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])  # Orijinal noktalar
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])  # Hedef noktalar
matrix_affine = cv2.getAffineTransform(pts1, pts2)
image_affine = cv2.warpAffine(image_rgb, matrix_affine, (w, h))

# 5Ô∏è‚É£ **Perspektif D√∂n√º≈ü√ºm√º (Perspective Transform)**
pts1_persp = np.float32([[100, 100], [400, 100], [100, 400], [400, 400]])
pts2_persp = np.float32([[50, 50], [450, 50], [100, 500], [400, 450]])
matrix_perspective = cv2.getPerspectiveTransform(pts1_persp, pts2_persp)
image_perspective = cv2.warpPerspective(image_rgb, matrix_perspective, (w, h))

# üîπ **T√ºm D√∂n√º≈ü√ºmleri G√∂rselle≈ütirme**
fig, axes = plt.subplots(2, 3, figsize=(20, 10))

# Orijinal G√∂r√ºnt√º
axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title("Orijinal G√∂r√ºnt√º")
axes[0, 0].axis("off")

# B√ºy√ºtme
axes[0, 1].imshow(image_resized_up)
axes[0, 1].set_title("B√ºy√ºt√ºlm√º≈ü G√∂r√ºnt√º (%150)")
axes[0, 1].axis("off")

# K√º√ß√ºltme
axes[0, 2].imshow(image_resized_down)
axes[0, 2].set_title("K√º√ß√ºlt√ºlm√º≈ü G√∂r√ºnt√º (%50)")
axes[0, 2].axis("off")

# D√∂nd√ºrme
axes[1, 0].imshow(image_rotated)
axes[1, 0].set_title("45 Derece D√∂nd√ºrme")
axes[1, 0].axis("off")

# Affine √áarpƒ±tma
axes[1, 1].imshow(image_affine)
axes[1, 1].set_title("Affine Transform (√áarpƒ±tma)")
axes[1, 1].axis("off")

# Perspektif D√∂n√º≈ü√ºm
axes[1, 2].imshow(image_perspective)
axes[1, 2].set_title("Perspektif D√∂n√º≈ü√ºm")
axes[1, 2].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# üìå G√∂r√ºnt√ºy√º y√ºkle ve gri tonlamaya √ßevir
image = cv2.imread("C:\\Goruntu\\rice_grain.jpg", cv2.IMREAD_GRAYSCALE) 

# √áekirdek (Kernel) tanƒ±mlama (5x5 boyutunda)
kernel = np.ones((5,5), np.uint8)

# Morfolojik ƒ∞≈ülemleri Uygula
image_erosion = cv2.erode(image, kernel, iterations=1)  # Erozyon
image_dilation = cv2.dilate(image, kernel, iterations=1)  # Geni≈üleme
image_opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)  # A√ßƒ±lma
image_closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)  # Kapanma
image_gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)  # Gradient
image_tophat = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel)  # Top Hat
image_blackhat = cv2.morphologyEx(image, cv2.MORPH_BLACKHAT, kernel)  # Black Hat

# üîπ **Jupyter Notebook i√ßin G√∂rselle≈ütirme**
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

# Orijinal G√∂r√ºnt√º
axes[0, 0].imshow(image, cmap="gray")
axes[0, 0].set_title("Orijinal G√∂r√ºnt√º")
axes[0, 0].axis("off")

# Erozyon
axes[0, 1].imshow(image_erosion, cmap="gray")
axes[0, 1].set_title("Erozyon (Erosion)")
axes[0, 1].axis("off")

# Geni≈üleme
axes[0, 2].imshow(image_dilation, cmap="gray")
axes[0, 2].set_title("Geni≈üleme (Dilation)")
axes[0, 2].axis("off")

# A√ßƒ±lma
axes[0, 3].imshow(image_opening, cmap="gray")
axes[0, 3].set_title("A√ßƒ±lma (Opening)")
axes[0, 3].axis("off")

# Kapanma
axes[1, 0].imshow(image_closing, cmap="gray")
axes[1, 0].set_title("Kapanma (Closing)")
axes[1, 0].axis("off")

# Gradient
axes[1, 1].imshow(image_gradient, cmap="gray")
axes[1, 1].set_title("Morphological Gradient")
axes[1, 1].axis("off")

# Top Hat
axes[1, 2].imshow(image_tophat, cmap="gray")
axes[1, 2].set_title("Top Hat")
axes[1, 2].axis("off")

# Black Hat
axes[1, 3].imshow(image_blackhat, cmap="gray")
axes[1, 3].set_title("Black Hat")
axes[1, 3].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# üìå G√∂r√ºnt√ºy√º y√ºkleme
image = cv2.imread("C:\\Goruntu\\rice_grain.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 1Ô∏è‚É£ **Gri tonlamaya √ßevirme**
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2Ô∏è‚É£ **G√ºr√ºlt√º azaltma i√ßin bulanƒ±kla≈ütƒ±rma (Gaussian Blur)**
blurred = cv2.GaussianBlur(gray, (5,5), 0)

# 3Ô∏è‚É£ **E≈üikleme (Thresholding)**
_, thresh = cv2.threshold(blurred, 125, 255,  cv2.THRESH_OTSU)

# 4Ô∏è‚É£ **Morfolojik A√ßƒ±lma (K√º√ß√ºk g√ºr√ºlt√ºleri temizleme)**
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=4)

# 5Ô∏è‚É£ **Arka plan maskesi (Geni≈üleme ile)**
sure_bg = cv2.dilate(opening, kernel, iterations=2)

# 6Ô∏è‚É£ **√ñn plan maskesi (Mesafe d√∂n√º≈ü√ºm√º ile)**
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
_, sure_fg = cv2.threshold(dist_transform, 0.3 * dist_transform.max(), 255, 0)

# 7Ô∏è‚É£ **Bilinmeyen b√∂lge belirleme (Arka plan - √ñn plan)**
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)

# 8Ô∏è‚É£ **Marker etiketleme**
_, markers = cv2.connectedComponents(sure_fg)

# **Arka planƒ± 1 deƒüil -1 yaparak Watershed algoritmasƒ± i√ßin ayarlƒ±yoruz**
markers = markers + 1
markers[unknown == 255] = 0

# 9Ô∏è‚É£ **Watershed Algoritmasƒ±nƒ± √áalƒ±≈ütƒ±r**
markers = cv2.watershed(image, markers)
image[markers == -1] = [255, 0, 0]  # Kenarlarƒ± kƒ±rmƒ±zƒ± renkte √ßiz

# üîπ **T√ºm Adƒ±mlarƒ± G√∂rselle≈ütirme**
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

# Orijinal G√∂r√ºnt√º
axes[0, 0].imshow(image_rgb)
axes[0, 0].set_title("Orijinal G√∂r√ºnt√º")
axes[0, 0].axis("off")

# Gri Tonlama
axes[0, 1].imshow(gray, cmap="gray")
axes[0, 1].set_title("Gri Tonlama")
axes[0, 1].axis("off")

# Threshold
axes[0, 2].imshow(thresh, cmap="gray")
axes[0, 2].set_title("Thresholding")
axes[0, 2].axis("off")

# A√ßƒ±lma (Morphology)
axes[0, 3].imshow(opening, cmap="gray")
axes[0, 3].set_title("Morfolojik A√ßƒ±lma")
axes[0, 3].axis("off")

# Arka Plan
axes[1, 0].imshow(sure_bg, cmap="gray")
axes[1, 0].set_title("Arka Plan Maskesi")
axes[1, 0].axis("off")

# √ñn Plan
axes[1, 1].imshow(sure_fg, cmap="gray")
axes[1, 1].set_title("√ñn Plan Maskesi")
axes[1, 1].axis("off")

# Watershed Sonucu
axes[1, 2].imshow(markers, cmap="jet")
axes[1, 2].set_title("Watershed Sonucu (Etiketler)")
axes[1, 2].axis("off")

# Segmentasyon Sonucu
axes[1, 3].imshow(image)
axes[1, 3].set_title("Watershed Sonu√ßlu G√∂r√ºnt√º")
axes[1, 3].axis("off")

plt.show()

num_labels, markers = cv2.connectedComponents(sure_fg)
nesne_sayisi = num_labels - 1  # Arka plan hari√ß ger√ßek nesne sayƒ±sƒ±
print(f"Tespit edilen nesne sayƒ±sƒ±: {nesne_sayisi}")
```
